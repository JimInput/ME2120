```python
import matplotlib.pyplot as plt
import numpy as np
```

Data:
$\begin{cases}
    x(0) = 0 \text{m} \\
    v(0) = 10 \text{m/s} \\
    m = 1 \text{kg} \\
    \mu_c = 0.5 \\
    b = 1 \text{kg/s} \\
    k = 4 \text{N/m} \\
\end{cases}$

```python
x0 = 0
v0 = 10
m = 1
mu_c = 0.5
b = 1
k = 4
g = 9.81
```

Case 1: 
A chair moving across rough terrain.

$x(t) = \begin{cases}
    x(0) + \dot{x}(0)t-\frac{1}{2}\mu_c g t^2 & \text{for  } 0 \leq t \leq \frac{\dot{x}(0)}{\mu_c g} \\
    x(0) + \frac{\dot{x}^2(0)}{2\mu_c g} & \text{for  } t \ge \frac{\dot{x}(0)}{\mu_c g}
\end{cases}$

```python
x = lambda t: x0 + v0*t-0.5*mu_c*g*t**2
v = lambda t: v0 - mu_c*g*t
t = np.linspace(0,3,51)

sol_x = x(t)
sol_v = v(t)

sol_x[sol_v<=0] = np.max(sol_x)

plt.plot(t, sol_x, 'b-')
plt.title("Position vs Time with Sliding Friction")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.grid()
plt.show()
```

Case 2: 
A block moving with viscous friction

$x(t) = x(0)+\frac{m}{b}\dot{x}(0)(1-e^{\frac{-b}{m}t})$

```python
x = lambda t: x0 + v0*m/b*(1-np.exp(-b/m*t))

sol_x = x(t)

plt.plot(t, sol_x, 'g-')
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.title("Position vs Time with Viscous Friction")
```

Case 3: 
A block in SHM by a spring.

$\omega_0 = \sqrt{\frac{k}{m}}$

$x(t) = x(0)\cos(\omega_0 t) + \frac{\dot{x}(0)}{\omega_0}\sin(\omega_0 t)$

```python
w0 = (k/m)**0.5
x = lambda t: x0*np.cos(w0*t) + v0/w0 * np.sin(w0*t)

plt.plot(t, x(t), 'r-')
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.title("Position vs Time with Simple Harmonic Motion")
```
