2. A chair slides across a waxed floor. Your goal is to show how far it moved and where it stopped. The only force slowing the chair down is the force of friction, so the kinematic equations are:

$\begin{cases}
    x(t) = x_0 + v_0 t - \frac{1}{2} \mu g t^2 \\
    v(t) = v_0 - \mu g t
\end{cases}$


but, these equations are only valid while $v(t) > 0$

$\begin{cases}
    \mu = 0.1 \\
    x_0 = 0 \text{m} \\
    v_0 = 3 \text{m/s} \\
    g = 9.81 \text{m/s/s}
\end{cases}$

a. Plot the position and velocity from 0-5 seconds

b. Update the plot to only show time, position, and velocity while $v(t)>0$

c. How far did the chair slide?

```python
mu = 0.1
x0 = 0
v0 = 3
g = 9.81

t = np.linspace(0,5,100)
```

```python
x = x0+v0*t-0.5*mu*g*t**2
v = v0 - mu * g * t
```

```python
plt.plot(t, x)
plt.xlabel("time (s)")
plt.ylabel("position (m)")
plt.title("position vs time from 0-5s")
```

```python
plt.xlabel("time (s)")
plt.ylabel("velocity (m/s)")
plt.title("velocity vs time for 0-5s")
plt.plot(t, v)
```

```python
stop_index = v > 0

plt.xlabel("time (s)")
plt.ylabel("position (m)")
plt.title("position vs time until chair stops")
plt.plot(t[stop_index], x[stop_index])
```

```python
plt.xlabel("time (s)")
plt.ylabel("velocity (m/s)")
plt.title("velocity vs time until chair stops")
plt.plot(t[stop_index], v[stop_index])
```

```python
print(f"The resting position of the chair is at {x[stop_index][-1]:.3f} meters.")
```
