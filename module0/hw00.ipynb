2. A chair slides across a waxed floor. Your goal is to show how far it moved and where it stopped. The only force slowing the chair down is the force of friction, so the kinematic equations are:

$\begin{cases}
    x(t) = x_0 + v_0 t - \frac{1}{2} \mu g t^2 \\
    v(t) = v_0 - \mu g t
\end{cases}$


but, these equations are only valid while $v(t) > 0$

$\begin{cases}
    \mu = 0.1 \\
    x_0 = 0 \text{m} \\
    v_0 = 3 \text{m/s} \\
    g = 9.81 \text{m/s/s}
\end{cases}$

a. Plot the position and velocity from 0-5 seconds

b. Update the plot to only show time, position, and velocity while $v(t)>0$

c. How far did the chair slide?

```python
mu = 0.1 # unitless
x0 = 0 # m
v0 = 3 # m/s
g = 9.81 # m/s/s

t = np.linspace(0,5,500)
```

```python
x = x0+v0*t-0.5*mu*g*t**2
v = v0 - mu * g * t
```

```python
plt.xlabel("time (s)")
plt.ylabel("position (m) / velocity (m/s)")
plt.title("position and velocity vs time naïve")
plt.plot(t, x, label="Chair Position (m)")
plt.plot(t, v, label="Chair Velocity (m/s)")
plt.legend()
```

```python
plt.xlabel("time (s)")
plt.ylabel("position (m) / velocity (m/s)")
plt.title("position vs time until chair stops")
plt.plot(t[v >= 0], x[v >= 0], label="Chair Position (m)")
plt.plot(t[v >= 0], v[v >= 0], label="Chair Velocity (m/s)")
plt.legend()
```

Now show plot of both from 0-5s 

```python
v[v<0] = 0
x[v==0] = np.max(x)
plt.plot(t,x, label='Chair Position (m)')
plt.plot(t,v, label='Chair Velocity (m/s)')
plt.legend()
plt.xlabel("time(s)")
plt.ylabel("position (m) / velocity (m/s)")
plt.title("position and velocity vs time from 0-5s")
```

```python
print(f"The resting position of the chair is at {np.max(x):.3f} meters.")
```
